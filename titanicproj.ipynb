```python
import numpy as numpy
import pandas as pandas
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
trainCSV = pandas.read_csv('../input/titanic/train.csv')
trainCSV.head()
```

```python
trainCSV.info()
```

```python
trainCSV_2 = trainCSV.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
trainCSV_2.head()
```

```python
model = SimpleImputer(strategy='mean')
```

```python
CSVnum = trainCSV_2.drop(['Embarked', 'Sex'], axis=1) 
```

```python
model.fit(CSVnum)
```

```python
model.statistics_
```

```python
dataFrameX = model.transform(CSVnum)  
dataFrameX[:5, :]
```

```python
CSVTransform = pandas.DataFrame(dataFrameX, columns=CSVnum.columns, index=CSVnum.index)
```

```python
CSVTransform.head()
```

```python
CSVTransform.info()
```

```python
dataFrameX = CSVTransform.drop('Survived', axis=1)
dataFrameY = CSVTransform['Survived'].copy()
```

```python
dataFrameX.head()
```

```python
dataFrameY
```

```python
trainDataframeX, testDataframeX, trainDataframeY, testDataframeY = train_test_split(dataFrameX, dataFrameY, test_size=0.20, random_state=42)
```

```python
LogReg = LogisticRegression(max_iter=200,random_state=42).fit(trainDataframeX, trainDataframeY) 
```

```python
LogReg.predict(testDataframeX)
```

```python
LogRegCV = LogisticRegressionCV(cv=5, max_iter=300).fit(trainDataframeX, trainDataframeY) # choosing cross validate = 5 fold
```

```python
LogRegCV.predict(testDataframeX)
```

```python
LogRegCV.score(testDataframeX, testDataframeY)
```

```python
OrdinalEncoder = OrdinalEncoder()
titanicSex = numpy.c_[trainCSV_2['Sex']] # It will give us 2-D array
```

```python
sexEncoded = OrdinalEncoder.fit_transform(titanicSex)
sexEncoded[:5]
```

```python
CSVTransform['Sex'] = sexEncoded
```

```python
CSVTransform.head()
```

```python
DataFrameNewX = CSVTransform.drop(['Survived'], axis=1)
DataFrameNewY = CSVTransform['Survived'].copy()
```

```python
trainDataFrameNewX, testDataFrameNewX, trainDataFrameNewY, testDataFrameNewY = train_test_split(DataFrameNewX, DataFrameNewY, test_size=0.30, random_state=42)
trainDataFrameNewX.head()
```

```python
LogReg = LogisticRegression(max_iter=200).fit(trainDataFrameNewX, trainDataFrameNewY)
LogReg.predict(testDataFrameNewX)
```

```python
LogReg.score(testDataFrameNewX, testDataFrameNewY)
```

```python
testCSV = pandas.read_csv('../input/titanic/test.csv')
testCSV.head()
```

```python
testCSV = testCSV.drop(['Name', 'Ticket','Cabin', 'Embarked'], axis=1)
```

```python
sexCatTestCSV = numpy.c_[testCSV['Sex']]
testCSVSexEncoded = OrdinalEncoder.fit_transform(sexCatTestCSV)
```

```python
testCSVSexEncoded[:5]
```

```python
testCSVNumberAttribute = testCSV.drop(['Sex', 'PassengerId'], axis=1)
testPassengerId = testCSV['PassengerId'].copy()
testCSVNumberAttribute
```

```python
testCSVImputer = model.fit_transform(testCSVNumberAttribute)
testCSVImputer
```

```python
testCSVImputer = pandas.DataFrame(testCSVImputer, columns=testCSVNumberAttribute.columns, index=testCSVNumberAttribute.index)
testCSVImputer.head()
```

```python
testCSVImputer['Sex'] = testCSVSexEncoded
testCSVImputer['PassengerId'] = testPassengerId
testCSVImputer = testCSVImputer[['PassengerId', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex']]
testCSVImputer.head()
```

```python
testSurvived = LogReg.predict(testCSVImputer)
testSurvived[:20]
```

```python
testCSVImputer['Survived'] = testSurvived
```

```python
testCSVImputer['Survived'] = testCSVImputer['Survived'].astype(int)
testCSVImputer.head()
```

```python
finalOutput = testCSVImputer.drop(['Pclass', 'Age', 'SibSp', 'Parch', 'Sex', 'Fare'], axis=1)
finalOutput.head()
```

```python
finalOutput.to_csv('titanicSurvivals.csv', index=False)
```
